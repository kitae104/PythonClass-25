# 딕셔너리 기초

딕셔너리는 `이름: 값` 형태로 정보를 저장하는 자료형입니다.

- 사전처럼 키와 값을 짝지어 저장합니다.
- 중괄호 `{}`로 만듭니다.
- 정보를 찾을 때 이름표(키)를 사용합니다.

```python
student = {"name": "민준", "grade": 5, "favorite": "축구"}
print(student)
print("이름:", student["name"])
print("좋아하는 것:", student["favorite"])
```

## 값 바꾸기와 새 정보 추가하기

딕셔너리도 나중에 내용을 바꾸거나 새 정보를 추가할 수 있습니다.

```python
game_character = {"name": "용감한 토끼", "level": 1}
game_character["level"] = 2
game_character["item"] = "당근 검"
print(game_character)
```

## 딕셔너리의 키와 값 보기

딕셔너리에 어떤 정보가 들어 있는지 하나씩 확인할 수도 있습니다.

```python
profile = {"이름": "지우", "취미": "그림 그리기", "좋아하는 색": "하늘색"}
for key, value in profile.items():
    print(key, ":", value)
```

## 재미있는 미니 예제

간단한 메뉴판을 딕셔너리로 만들어 봅시다.

```python
menu = {"핫도그": 2500, "슬러시": 1500, "쿠키": 1000}
print("핫도그 가격:", menu["핫도그"], "원")
menu["쿠키"] = 1200  # 가격 수정
print("새로운 메뉴판:", menu)
```
