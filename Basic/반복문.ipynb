# 반복문 기초

반복문은 같은 일을 여러 번 할 때 사용합니다.

- `for`문은 정해진 횟수나 목록을 반복할 때 사용합니다.
- `while`문은 조건이 참인 동안 반복합니다.

```python
for count in range(1, 6):
    print(count, "번 점프!")
```

## 문자열이나 리스트도 반복할 수 있어요

글자 하나씩, 또는 리스트의 값 하나씩 꺼내서 볼 수 있습니다.

```python
characters = ["마법사", "기사", "궁수"]
for character in characters:
    print(character, "출발!")
```

## while문으로 카운트하기

조건이 맞는 동안 계속 반복합니다.

```python
energy = 3
while energy > 0:
    print("달린다! 남은 에너지:", energy)
    energy -= 1  # 1씩 줄입니다.
print("휴식 시간!")
```

## 재미있는 미니 예제

별을 하나씩 늘려 가며 작은 삼각형을 만들어 봅시다.

```python
for line in range(1, 6):
    print("*" * line)
```
