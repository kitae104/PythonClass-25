# 종합 예제: 딕셔너리 + 리스트 + if + for + input

이 노트북은 파이썬 기초 문법을 함께 연습할 수 있도록 만든 **종합 예제 2개**를 포함합니다.

각 예제는 다음 요소를 모두 사용합니다.
- 리스트(list)
- 딕셔너리(dict)
- 조건문(if)
- 반복문(for)
- 사용자 입력(input)

## 예제 1) 장바구니 계산기

사용자가 과일 이름을 쉼표(,)로 입력하면,
- 입력값을 리스트로 변환하고
- 과일 가격 딕셔너리에서 가격을 찾은 뒤
- if로 유효한 상품 여부를 판단하고
- for로 총합을 계산합니다.

```python
# 과일 가격표 (딕셔너리)
price_table = {
    "사과": 1200,
    "바나나": 800,
    "오렌지": 1500,
    "포도": 2000
}

raw = input("구매할 과일 이름을 쉼표(,)로 구분해 입력하세요: ")
# input 결과를 리스트로 변환
cart = [item.strip() for item in raw.split(",") if item.strip()]

total = 0
unknown_items = []

for fruit in cart:
    if fruit in price_table:
        total += price_table[fruit]
    else:
        unknown_items.append(fruit)

print("\n[입력한 장바구니]", cart)
print("총 결제 금액:", total, "원")

if unknown_items:
    print("가격표에 없는 상품:", unknown_items)
else:
    print("모든 상품이 정상적으로 계산되었습니다.")
```

## 예제 2) 간단 퀴즈 채점기

문제와 정답을 딕셔너리로 관리하고,
for 반복문으로 사용자 답안을 입력받아 리스트에 저장한 뒤,
if 조건문으로 정오답을 판별해 점수를 계산합니다.

```python
# 문제:정답 (딕셔너리)
quiz = {
    "파이썬에서 리스트를 만드는 기호는?": "[]",
    "파이썬에서 딕셔너리를 만드는 기호는?": "{}",
    "반복문 키워드 중 하나는?": "for"
}

questions = list(quiz.keys())      # 딕셔너리 키를 리스트로 변환
user_answers = []                  # 사용자 답 저장용 리스트
score = 0

for q in questions:
    ans = input(f"문제: {q} ")
    user_answers.append(ans.strip())

print("\n[채점 결과]")
for i in range(len(questions)):
    q = questions[i]
    correct = quiz[q]
    mine = user_answers[i]

    if mine == correct:
        print(f"{i+1}번 정답 (내 답: {mine})")
        score += 1
    else:
        print(f"{i+1}번 오답 (내 답: {mine}, 정답: {correct})")

print(f"최종 점수: {score}/{len(questions)}")

if score == len(questions):
    print("완벽합니다!")
elif score >= 2:
    print("좋아요! 조금만 더 연습해요.")
else:
    print("기초 문법을 한 번 더 복습해보세요.")
```

## 정리

- 딕셔너리는 "기준표(가격표, 정답표)"로 사용하기 좋습니다.
- 리스트는 여러 입력값(장바구니 항목, 사용자 답안)을 저장하기 좋습니다.
- if는 분기 처리(정상/예외, 정답/오답)에 사용합니다.
- for는 목록을 순회하면서 누적 계산/채점을 할 때 유용합니다.
- input은 사용자와 상호작용하는 프로그램의 시작점입니다.
