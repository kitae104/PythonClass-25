# 함수 기초

함수는 자주 쓰는 코드를 이름 붙여 묶어 둔 것입니다.

- `def`로 함수를 만듭니다.
- 필요할 때 함수 이름을 써서 실행합니다.
- 입력값을 받아서 결과를 돌려줄 수도 있습니다.

```python
def say_hello():
    print("안녕하세요! 오늘도 파이썬 탐험 시작!")

say_hello()
```

## 값을 받아서 사용하는 함수

함수에 필요한 정보를 전달하면 더 다양한 일을 시킬 수 있습니다.

```python
def cheer(name):
    print(name, "파이팅!")

cheer("수아")
cheer("지호")
```

## 값을 계산해서 돌려주는 함수

`return`을 사용하면 계산 결과를 함수 밖으로 보낼 수 있습니다.

```python
def add_points(score1, score2):
    total = score1 + score2
    return total

result = add_points(15, 20)
print("총 점수:", result)
```

## 재미있는 미니 예제

간단한 메뉴 추천 함수를 만들어 봅시다.

```python
def recommend_menu(weather):
    if weather == "비":
        return "따끈한 우동"
    return "시원한 냉면"

print("비 오는 날 추천:", recommend_menu("비"))
print("맑은 날 추천:", recommend_menu("맑음"))
```

📌 문제 설명

학생들의 점수가 담긴 리스트가 있을 때
👉 평균 점수를 계산하고
👉 합격자(60점 이상)만 출력하는 프로그램을 작성하시오.

```python
# 학생 데이터 (리스트 + 딕셔너리)
students = [
    {"name": "김철수", "score": 85},
    {"name": "이영희", "score": 58},
    {"name": "박민수", "score": 72}
]

# 평균 점수 계산 함수
def get_average(student_list):
    total = 0
    for student in student_list:
        total += student["score"]
    return total / len(student_list)

# 합격자 출력 함수
def print_pass_students(student_list):
    for student in student_list:
        if student["score"] >= 60:
            print(student["name"], "합격")

# 실행
avg = get_average(students)
print("평균 점수:", avg)

print("합격자:")
print_pass_students(students)
```

📌 문제 설명

장바구니에 담긴 상품들의
👉 총 금액을 계산하는 함수를 만들고
👉 5만원 이상이면 할인 적용(10%)

```python
# 장바구니 (리스트 + 딕셔너리)
cart = [
    {"item": "노트북", "price": 1000000, "qty": 1},
    {"item": "마우스", "price": 20000, "qty": 2},
    {"item": "키보드", "price": 50000, "qty": 1}
]

# 총 금액 계산 함수
def get_total(cart_list):
    total = 0
    for product in cart_list:
        total += product["price"] * product["qty"]
    return total

# 할인 적용 함수
def apply_discount(total):
    if total >= 50000:
        return total * 0.9
    return total

# 실행
total_price = get_total(cart)
final_price = apply_discount(total_price)

print("총 금액:", total_price)
print("할인 적용 금액:", final_price)
```
