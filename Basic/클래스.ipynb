# 클래스 기초

클래스는 비슷한 특징과 기능을 가진 대상을 만들기 위한 설계도입니다.

- 클래스는 `class`로 만듭니다.
- 클래스로 만든 실제 대상을 객체라고 합니다.
- 객체는 데이터와 기능을 함께 가질 수 있습니다.

```python
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        print(self.name, "멍멍!")

dog1 = Dog("초코")
dog1.bark()
```

## 객체마다 다른 값을 가질 수 있어요

같은 클래스에서 여러 객체를 만들면 이름이나 점수 같은 값이 달라질 수 있습니다.

```python
class GameCharacter:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def introduce(self):
        print("나는", self.name, "이고, 힘은", self.power, "이야!")

hero = GameCharacter("별빛 기사", 80)
wizard = GameCharacter("구름 마법사", 95)
hero.introduce()
wizard.introduce()
```

## 클래스의 기능 만들기

클래스 안에 함수를 넣으면 객체가 할 수 있는 행동을 만들 수 있습니다.

```python
class Robot:
    def __init__(self, name):
        self.name = name
        self.energy = 3

    def work(self):
        if self.energy > 0:
            self.energy -= 1
            print(self.name, "이(가) 청소 중! 남은 에너지:", self.energy)
        else:
            print(self.name, "은(는) 충전이 필요해요.")

robot = Robot("반짝봇")
robot.work()
robot.work()
robot.work()
robot.work()
```
