# 리스트 기초

리스트는 여러 값을 순서대로 담는 상자입니다.

- 대괄호 `[]`로 만듭니다.
- 순서가 있어서 몇 번째 값인지 확인할 수 있습니다.
- 값을 추가하거나 수정할 수 있습니다.

```python
snacks = ["떡볶이", "김밥", "어묵"]
print("오늘의 간식 목록:", snacks)
print("첫 번째 간식:", snacks[0])
print("마지막 간식:", snacks[-1])
```

## 리스트 값 바꾸기와 추가하기

리스트는 만들어 놓은 뒤에도 내용을 바꿀 수 있습니다.

```python
animals = ["고양이", "강아지", "토끼"]
animals[1] = "판다"  # 두 번째 값을 바꿉니다.
animals.append("햄스터")  # 맨 뒤에 새 값을 추가합니다.
print(animals)
```

## 리스트와 반복문

리스트에 들어 있는 값을 하나씩 꺼내면서 사용할 수 있습니다.

```python
missions = ["양치하기", "숙제하기", "물 마시기"]
for mission in missions:
    print(mission, "완료!")
```

## 재미있는 미니 예제

주사위를 세 번 굴린 결과를 리스트에 담아 보고 합계도 구해 봅시다.

```python
dice_scores = [3, 5, 2]
print("주사위 결과:", dice_scores)
print("점수 합계:", sum(dice_scores))
print("가장 큰 수:", max(dice_scores))
```
