```python
import pandas as pd

df = pd.read_csv("bmi_data.csv", encoding="utf-8")
print(df)
names = df["이름"].tolist()
heights = df["키"].astype(float).tolist()
weights = df["몸무게"].astype(float).tolist()

print("이름:", names)
print("키:", heights) 
print("몸무게:", weights)
```

```python
def calculate_bmi(height_cm, weight_kg):
  height_m = height_cm / 100  # cm를 m로 변환
  bmi = weight_kg / (height_m ** 2)
  return bmi
```

```python
def decision(bmi):
  if bmi < 18.5:
    return "저체중"
  elif 18.5 <= bmi < 23:
    return "정상"
  elif 23 <= bmi < 25:
    return "고체중"
  else:
    return "비만"
```

```python
for i in range(len(names)):
  bmi = calculate_bmi(heights[i], weights[i])
  status = decision(bmi)
  print(f"{names[i]}의 BMI: {bmi:.2f}, 상태: {status}")
```

```python
# BMI 계산 및 결과 출력
for name, height, weight in zip(names, heights, weights):
    bmi = calculate_bmi(height, weight)
    status = decision(bmi)
    print(f"{name}의 BMI: {bmi:.2f}, 상태: {status}")
```
